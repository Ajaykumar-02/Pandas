## Diffrent types of Series in Pandas

```python
lables=['a','b','c']
my_list=[10,20,30]
arr=np.array([10,20,30])
d = {1:10,2:20,3:30}
```

```python
pd.Series(my_list) #by Defult lables = 0,1,2
```

```python
pd.Series(my_list,index=lables) #lables define using index
```

```python
pd.Series(arr)
```

```python
pd.Series(arr,index=lables)
```

```python
pd.Series(d) # in dict key is a lables
```
